```python
from sklearn.preprocessing import StandardScaler
```

```python
#Making whole data into one format
sc= StandardScaler()
sc.fit(gas_turbines)
gas_turbines_stndrd = sc.transform(gas_turbines)
gas_turbines_stndrd
```

```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import backend as K
```

```python
def rmse(y_true, y_pred):
    return K.sqrt(K.mean(K.square(y_pred - y_true)))
#(RMSE) is a standard way to measure the error of a model in predicting quantitative data. RMSE values between 0.2 and 0.5 shows that the model can relatively predict the data accurately
```

```python
# split into input (X) and output (Y) variables
X = gas_turbines_stndrd[:,0:10]
Y = gas_turbines_stndrd[:,10]
Y
```

```python
tf.random.set_seed(42)
# declare a model. It will be a simple one, having just three hidden layers of 256, 256, and 128 units. Feel free to experiment with these, as thereâ€™s no right or wrong way to set up a neural network. These layers are then followed by an output layer of one node, since youâ€™re predicting a numerical value.
#Rectified Linear Unit (ReLU) - Graphically, the ReLU function is composed of two linear pieces to account for non-linearities. A function is non-linear if the slope isnâ€™t constant. So, the ReLU function is non-linear around 0, but the slope is always either 0 (for negative inputs) or 1 (for positive inputs).
model = Sequential([
    Dense(256, activation='relu'),
    Dense(256, activation='relu'),
    Dense(128, activation='relu'),
    Dense(1)
])
#compile a model using the RMSE as a way to keep track of the loss and as an evaluation metric, and youâ€™ll optimize the model using the Adam optimizer.
model.compile(
    loss=rmse,
    optimizer=Adam(),
    metrics=[rmse]
)
#train the model on the data for 100 epochs
model.fit(X, Y, epochs=100)
```
