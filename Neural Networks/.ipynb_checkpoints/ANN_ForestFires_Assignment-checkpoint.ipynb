```python
# Create your first MLP in Keras
from tensorflow import keras
from keras.models import Sequential
from keras.layers import Dense
import numpy
```

```python
from pandas import read_csv
```

```python
forestfires = read_csv('forestfires.csv')
forestfires
```

```python
#Delete columns 'month', 'day' as it already exists
forestfires.drop(['month', 'day'], axis=1, inplace=True)
forestfires
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
forestfires_stndrd = forestfires[['FFMC', 'DMC', 'DC', 'ISI', 'temp', 'RH', 'wind', 'rain', 'area']]
forestfires_stndrd
```

```python
#Making whole data into one format
sc= StandardScaler()
sc.fit(forestfires_stndrd)
forestfires_stndrd = sc.transform(forestfires_stndrd)
forestfires_stndrd
```

```python
forestfires['FFMC'] = forestfires_stndrd[:,:1]
```

```python
forestfires['DMC'] = forestfires_stndrd[:,1:2]
forestfires['DC'] = forestfires_stndrd[:,2:3]
forestfires['ISI'] = forestfires_stndrd[:,3:4]
forestfires['temp'] = forestfires_stndrd[:,4:5]
forestfires['RH'] = forestfires_stndrd[:,5:6]
forestfires['wind'] = forestfires_stndrd[:,6:7]
forestfires['rain'] = forestfires_stndrd[:,7:8]
forestfires['area'] = forestfires_stndrd[:,8:9]
forestfires
```

```python
forestfires['size_category'] = forestfires['size_category'].map({'large': 1, 'small': 0})
forestfires['size_category']
```

```python
# fix random seed for reproducibility
seed = 7
numpy.random.seed(seed)
# split into input (X) and output (Y) variables
array = forestfires.values
X = array[:,0:28]
Y = array[:,28]
Y
```

```python
forestfires['size_category'].unique()
```

```python
X
```

```python
# declare a model. It will be a simple one, having just three hidden layers of 12 and 8 units. Feel free to experiment with these, as thereâ€™s no right or wrong way to set up a neural network. These layers are then followed by an output layer of one node, since youâ€™re predicting a numerical value.
#Sigmoid (Logistic) - a common S-shaped curve, output of the function is centered at 0.5 with a range from 0 to 1
#Rectified Linear Unit (ReLU) - Graphically, the ReLU function is composed of two linear pieces to account for non-linearities. A function is non-linear if the slope isnâ€™t constant. So, the ReLU function is non-linear around 0, but the slope is always either 0 (for negative inputs) or 1 (for positive inputs).
model = Sequential()
model.add(Dense(12, input_dim=28, kernel_initializer='uniform', activation='relu'))
model.add(Dense(8, kernel_initializer='uniform', activation='relu'))
model.add(Dense(1, kernel_initializer='uniform', activation='sigmoid'))
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
from keras.callbacks import History 
history = History()
```

```python
X_convert = numpy.asarray(X).astype(numpy.float32)
X_convert
```

```python
# Fit the model
model.fit(X_convert, Y, validation_split=0.33, epochs=150, batch_size=10,callbacks=[history])
```

```python
# evaluate the model
scores = model.evaluate(X_convert, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# Visualize training history

# list all data in history
model.history.history.keys()
```

```python
# summarize history for accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
# summarize history for loss
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'test'], loc='upper left')
plt.show()
```
