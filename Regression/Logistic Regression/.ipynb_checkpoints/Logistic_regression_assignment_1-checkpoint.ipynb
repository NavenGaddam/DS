```python
#balance, duration columns have NaN values replace them with mean 
Check_outliers['balance'].fillna(Check_outliers['balance'].mean(), inplace=True)
Check_outliers['duration'].fillna(Check_outliers['duration'].mean(), inplace=True)
```

```python
#Logistic regression and fit the model
classifier = LogisticRegression()
```

```python
classifier.fit(Check_outliers,Y)
```

```python
#Predict for X dataset
y_pred = classifier.predict(Check_outliers)
y_pred
```

```python
y_pred.size
```

```python
y_pred_df= pd.DataFrame({'actual': Y,
                         'predicted_prob': y_pred}) # as y_pred is array, we convert them
y_pred_df
```

```python
# Confusion Matrix for the model accuracy

confusion_matrix = confusion_matrix(Y,y_pred)
print (confusion_matrix)#we can also use pd.crosstab(actual,y_pred_df)
```

```python
pd.crosstab(Y,y_pred_df.predicted_prob)
```

```python
#Classification report
#Precision: When it predicts yes, how often is it correct
#"Recall" : When it's actually yes, how often does it predict yes
#Accuracy: Overall, how often is the classifier correct
#F1 score — What percent of positive predictions were correct - the best score is 1.0 and the worst is 0.0
print(classification_report(Y,y_pred)) 
```

```python
#Classification report -D

print(classification_report(Y,y_pred)) #? macro avg, weighted avg
```

```python
#Receiver Operating Characteristic(ROC) curve is a plot of the true positive rate against the false positive rate. It shows the tradeoff between sensitivity and specificity.
fpr, tpr, thresholds = roc_curve(Y, classifier.predict_proba (Check_outliers)[:,1])
```
